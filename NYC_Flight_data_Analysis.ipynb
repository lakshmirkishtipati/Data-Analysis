# Working with Missing Data

### 1) Identify Missing Data

All the Columns whose having True count has Null values

```python
def missing_values():
    missing_values = flight_df.isna()
    for column in missing_values.columns.values.tolist():
        print(column)
        print(missing_values[column].value_counts())
        print('\n')
missing_values()
```

## Based on the summary above, each column has 336776 rows of data and six columns containing missing data:

    1. "dep_time"       : 8255 missing values

    2. "dep_delay"      : 8255 missing values

    3. "arr_time"       : 8713 missing values

    4. "arr_delay"      : 9430 missing values

    5. "tailnum"        : 2512 missing values

    6. "air_time"       : 9430 missing values

### Deal with Missing Data

1) dep_delay

    i) Replacing missing values with mean value of dep_delay based on group by of tailnum origin and carrier.
    ii) Replacing remaing missing values with mean value of dep_delay based on group by origin and carrier
    

```python
flight_df['dep_delay'].fillna(flight_df.groupby(['tailnum','origin','carrier'])['dep_delay'].transform('mean'),inplace=True)
flight_df['dep_delay'].fillna(flight_df.groupby(['carrier','origin'])['dep_delay'].transform('mean'),inplace=True)
flight_df['dep_delay'] = flight_df['dep_delay'].round()
```

# Updating sched_dep_time with date time format

```python
flight_df['dep_datetime'] = flight_df['year'].astype('str')+'-'+flight_df['month'].astype('str').apply(lambda x: x.zfill(2))+'-'+flight_df['day'].astype('str').apply(lambda y:y.zfill(2))+'.'+flight_df['sched_dep_time'].astype('str').apply(lambda z:z.zfill(4))

flight_df['sched_dep_time'] = pd.to_datetime(pd.Series(flight_df['dep_datetime']),format="%Y-%m-%d.%H%M")
print(flight_df['sched_dep_time'].head(5))
```

2) Dep_time : Filling Missing values and Correcting dep_time value with proper values(sched_dep_values+dep_delay).

```python
# filling dep_time values with scheduled_dep_time + dep_delay
flight_df['dep_delay_delta'] = pd.to_timedelta(flight_df['dep_delay'],unit='m')
flight_df['dep_time'] = flight_df['sched_dep_time']+flight_df['dep_delay_delta']

# Correcting dep_time value with proper values(sched_dep_values+dep_delay).
def adjust_sch_time(row):
    return row[['dep_time','sched_dep_time']]-pd.DateOffset(1)
flight_df.loc[flight_df['dep_time'].dt.day!=flight_df['day'],['dep_time','sched_dep_time']]=flight_df.loc[flight_df['dep_time'].dt.day!=flight_df['day']].apply(adjust_sch_time, axis=1)
print(flight_df[['dep_time','sched_dep_time']].head())
```

3,4) Similary updating airival delay and arrival time

```python
flight_df['arr_delay'].fillna(flight_df.groupby(['carrier','tailnum','origin','dest'])['arr_delay'].transform('mean'),inplace=True)
flight_df['arr_delay'].fillna(flight_df.groupby(['carrier','origin','dest'])['arr_delay'].transform('mean'),inplace=True)
flight_df.dropna(subset = ['arr_delay'],axis=0,inplace=True)
flight_df['arr_delay'] = flight_df['arr_delay'].round()
```

```python
flight_df['arr_datetime'] = flight_df['year'].astype('str')+'-'+flight_df['month'].astype('str').apply(lambda x: x.zfill(2))+'-'+flight_df['day'].astype('str').apply(lambda y:y.zfill(2))+'.'+flight_df['sched_arr_time'].astype('str').apply(lambda z:z.zfill(4))
flight_df['sched_arr_time'] = pd.to_datetime(pd.Series(flight_df['arr_datetime']),format="%Y-%m-%d.%H%M")
```

```python
def adjust_arr_time(row):
    if row['sched_arr_time']<row['sched_dep_time']:
        return row['sched_arr_time']+pd.DateOffset(1)
    elif (row['sched_arr_time']-row['sched_dep_time']).days>0:
        return row['sched_arr_time']-pd.DateOffset(1)
    else:
        return row['sched_arr_time']
flight_df['sched_arr_time'] = flight_df.apply(adjust_arr_time,axis=1)
print(flight_df['sched_arr_time'].head())
```

```python
flight_df['arr_delay_delta'] = pd.to_timedelta(flight_df['arr_delay'],unit='m')

flight_df['arr_time'] = flight_df['sched_arr_time']+flight_df['arr_delay_delta']
```

```python
def adjust_arr_time(row):
    return row['arr_time']+pd.DateOffset(1)
flight_df.loc[flight_df['arr_time']<flight_df['dep_time'],'arr_time']=flight_df.loc[flight_df['arr_time']<flight_df['dep_time']].apply(adjust_arr_time, axis=1)
print(flight_df[['arr_time','arr_delay','sched_arr_time']].head(8))
```

5) Airtime Replacing Null values with mean of air time For same tailnum, origin, destination.

```python
flight_df['air_time_actual'] = flight_df['air_time']
flight_df['air_time'].fillna(flight_df.groupby(['carrier','origin','dest'])['air_time_actual'].transform('mean'),inplace=True)
flight_df['air_time'].fillna(flight_df.groupby(['origin','dest'])['air_time_actual'].transform('mean'),inplace=True)
```
